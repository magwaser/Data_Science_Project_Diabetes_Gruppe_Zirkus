### 4.1 The multicollinearity variance inflation factor (VIF) <a id='4.1'></a>
... can be calculated to figure out the
standard of tolerance and assess the degree of collinearity

```python
# compute the vif for all given features
def compute_vif(considered_features):
    
    X = data[considered_features]
    # the calculation of variance inflation requires a constant
    X['intercept'] = 1
    
    # create dataframe to store vif values
    vif = pd.DataFrame()
    vif["Variable"] = X.columns
    vif["VIF"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
    vif = vif[vif['Variable']!='intercept']
    return vif

# von: https://towardsdatascience.com/targeting-multicollinearity-with-python-3bd3b4088d0b
```

```python
# features to consider removing
considered_features = ['Pregnancies','Glucose','BloodPressure', 'SkinThickness', 'Insulin', 'BMI', 'DiabetesPedigreeFunction', 'Age']


# compute vif 
compute_vif(considered_features).sort_values('VIF', ascending=False)
```

    -> Vermutlich keine multi-collinearity vorhanden

### 6.2 Hyperparameter tuning function for RF <a id = '6.2'></a>

```python
def RF_hyper(X_test, y_test):
    param_grid = {
    'n_estimators': [25, 50, 100, 150],
    'max_features': ['sqrt', 'log2', None],
    'max_depth': [3, 6, 9],
    'max_leaf_nodes': [3, 6, 9],
    }
    random_search = RandomizedSearchCV(RandomForestClassifier(),param_grid, random_state = 69)
    random_search.fit(X_test, y_test)
    
    best_para = random_search.best_params_
    max_depth, max_features, max_leaf_nodes, n_estimators = best_para['max_depth'],best_para['max_features'],best_para['max_leaf_nodes'],best_para['n_estimators']
    
    return max_depth, max_features, max_leaf_nodes, n_estimators
    
```

```python
y_1d_model = column_or_1d(y_model)
max_depth, max_features, max_leaf_nodes, n_estimators = RF_hyper(X_model, y_1d_model)
```

## 9. Performance evaluation <a id ='9.0'></a>

### 9.1 eval_Performance Funktionen <a id ='9.1'></a>

```python
# Funktion fÃ¼r die Maschine Learning Evaluation
def eval_Performance(y_eval, X_eval, clf):
    '''
    y_eval = y_test
    X_eval = X_test
    clf = Maschine Learning Tool. (E.G. Log. Regression)
    '''
    
    # predict y based on our Maschine learning Model and create confusion matrix
    y_pred = clf.predict(X_eval) 
    y_pred_proba = clf.predict_proba(X_eval)[:, 1] 
    tn, fp, fn, tp = confusion_matrix(y_eval, y_pred).ravel() # true negatives & positives, false neg. & pos. 

    # Evaluation
    accuracy  = accuracy_score(y_eval, y_pred)
    precision = precision_score(y_eval, y_pred)
    specificity = tn / (tn + fp)
    recall    = recall_score(y_eval, y_pred)
    f1        = f1_score(y_eval, y_pred)
    fp_rates, tp_rates, _ = roc_curve(y_eval, y_pred_proba) # false-positive and true-positive rate

    # Calculate the area under the roc curve
    roc_auc = auc(fp_rates, tp_rates)

    return tp,fp,tn,fn,accuracy, specificity, precision, recall, f1, roc_auc
```

### 9.2 K-Fold performance function <a id ='9.2'></a>

Die flolgende Funktion berechnet die Performance fÃ¼r alle 4 CLFs bei einem gegebenem Datensatz

```python
def K_fold_performance(X, y, K):
    '''
    X = feature
    y = labels
    K = hyperparameter for KNN
    '''
    # implementiere den Performance DatdaFrame
    df_performance = pd.DataFrame(columns = ['tp','fp','tn','fn','accuracy', 'specificity','precision', 'recall', 'f1', 'roc_auc'] )
    
    # implementiere HilfsdatasÃ¤tze, fÃ¼r die Cross Validation
    df_performance_SVM_test = pd.DataFrame(columns = ['tp','fp','tn','fn','accuracy', 'specificity','precision', 'recall', 'f1', 'roc_auc'] )
    df_performance_SVM_train = pd.DataFrame(columns = ['tp','fp','tn','fn','accuracy', 'specificity','precision', 'recall', 'f1', 'roc_auc'] )

    df_performance_LR_test = pd.DataFrame(columns = ['tp','fp','tn','fn','accuracy', 'specificity','precision', 'recall', 'f1', 'roc_auc'] )
    df_performance_LR_train = pd.DataFrame(columns = ['tp','fp','tn','fn','accuracy', 'specificity','precision', 'recall', 'f1', 'roc_auc'] )

    df_performance_RF_test = pd.DataFrame(columns = ['tp','fp','tn','fn','accuracy', 'specificity','precision', 'recall', 'f1', 'roc_auc'] )
    df_performance_RF_train = pd.DataFrame(columns = ['tp','fp','tn','fn','accuracy', 'specificity','precision', 'recall', 'f1', 'roc_auc'] )

    df_performance_KNN_test = pd.DataFrame(columns = ['tp','fp','tn','fn','accuracy', 'specificity','precision', 'recall', 'f1', 'roc_auc'] )
    df_performance_KNN_train = pd.DataFrame(columns = ['tp','fp','tn','fn','accuracy', 'specificity','precision', 'recall', 'f1', 'roc_auc'] )
    
    # Implement a 5-fold stratified crossvalidation loop with relevant data standardization.

    # Perform a 5-fold crossvalidation - prepare the splits
    n_splits = 5
    skf      = StratifiedKFold(n_splits=n_splits)

    # counter to save our performance metrics for each crossvalidation fold
    fold = 0
    
    # Maschine learning tools
    
    clf_SVM = svm.SVC(probability=True, kernel = 'linear', random_state = 1) # Support Vector Maschine

    clf_LR = LogisticRegression(random_state=1)

    clf_KNN = KNeighborsClassifier(n_neighbors= K) # K-Nearest Neighbors
    # random_state, so we have the same results. Better for this groupproject, otherwise we would have a little chaos in our group.
    
    # prepare a list to save our clf models for later.
    classi = []

    # Loop over all splits
    for train_index, test_index in skf.split(X, y):

        # Get the relevant subsets for training and testing
        X_test  = X.iloc[test_index]
        y_test  = y.iloc[test_index]
        X_train = X.iloc[train_index]
        y_train = y.iloc[train_index]

        #transform y to an 1d array (better for sklearn)
        y_test = column_or_1d(y_test, warn=False)
        y_train = column_or_1d(y_train, warn=False)


        # Standardize the numerical features using training set statistics
        sc = StandardScaler()
        X_train_sc = sc.fit_transform(X_train)
        X_test_sc  = sc.transform(X_test)

        # Randomforest, but with hyperparameter tuning
        max_depth, max_features, max_leaf_nodes, n_estimators = RF_hyper(X_model, y_model_1d) # we defined this function in chapter ...
        clf_RF = RandomForestClassifier(random_state=1, max_depth = max_depth, max_features = max_features, max_leaf_nodes = max_leaf_nodes, n_estimators = n_estimators)
        
        # Creat prediction models and fit them to the training data

        # Logistic regression
        clf_LR.fit(X_train_sc, y_train)

        # Random forest
        clf_RF.fit(X_train_sc, y_train)

        #Support Vector Maschine
        clf_SVM.fit(X_train_sc, y_train)

        # K Nearest Neighbors
        clf_KNN.fit(X_train_sc, y_train)
        
        # save each clf for each fold. 
        classi.append(clf_SVM)
        classi.append(clf_LR)
        classi.append(clf_KNN)
        classi.append(clf_RF)

        # save the evaluation metrics in our aid-dataframes. (For each Fold)
        df_performance_SVM_test.loc[fold, : ] = eval_Performance(y_test, X_test_sc, clf_SVM)
        df_performance_SVM_train.loc[fold, : ] = eval_Performance(y_train, X_train_sc, clf_SVM)

        df_performance_LR_test.loc[fold, : ] = eval_Performance(y_test, X_test_sc, clf_LR)
        df_performance_LR_train.loc[fold, : ] = eval_Performance(y_train, X_train_sc, clf_LR)

        df_performance_RF_test.loc[fold, : ] = eval_Performance(y_test, X_test_sc, clf_RF)
        df_performance_RF_train.loc[fold, : ] = eval_Performance(y_train, X_train_sc, clf_RF)

        df_performance_KNN_test.loc[fold, : ] = eval_Performance(y_test, X_test_sc, clf_KNN)
        df_performance_KNN_train.loc[fold, : ] = eval_Performance(y_train, X_train_sc, clf_KNN)

        # increase counter for folds
        fold += 1
    
    # get the mean evaluation metrics over all folds and save them in another dataframe
    df_performance.loc['SVM (test)',:] = column_or_1d(df_performance_SVM_test.mean().round(3))
    df_performance.loc['SVM (train)',:] = column_or_1d(df_performance_SVM_train.mean().round(3))

    df_performance.loc['LR (test)',:] = column_or_1d(df_performance_LR_test.mean().round(3))
    df_performance.loc['LR (train)',:] = column_or_1d(df_performance_LR_train.mean().round(3))

    df_performance.loc['KNN (test)',:] = column_or_1d(df_performance_KNN_test.mean().round(3))
    df_performance.loc['KNN (train)',:] = column_or_1d(df_performance_KNN_train.mean().round(3))

    df_performance.loc['RF (test)',:] = column_or_1d(df_performance_RF_test.mean().round(3))
    df_performance.loc['RF (train)',:] = column_or_1d(df_performance_RF_train.mean().round(3))
    
    #return all calculated tables and classifiers
    return df_performance, classi, df_performance_SVM_test, df_performance_LR_test, df_performance_KNN_test, df_performance_RF_test
```

## 9.3 Evaluation of binary classifiers <a id = '9.3'></a>

Classification tasks are very common in data science applications and a large variety of evaluation metrics exist. For a data scientist it is important to understand how to calculate these evaluation metrics, but also to know when to use which of these. The basis for most evaluation metrics is the confusion matrix: 



#### 1. Accuracy

Accuracy is the most straightforward metric for classification. It measures the proportion of correct predictions out of the total number of predictions made.

\begin{align*}
   \text{Accuray} &= \frac{TP+TN}{TP + TN + FP + FN}
\end{align*}
#### 2. Precision

Precision, also known as positive predictive value, measures the proportion of true positive predictions out of all positive predictions made. It evaluates the model's ability to correctly identify positive instances.

\begin{align*}
   \text{Precision} &= \frac{TP}{TP + FP}
\end{align*}

#### 3. Recall

Recall, also known as sensitivity or true positive rate, measures the proportion of true positive predictions out of all actual positive instances. It evaluates the model's ability to identify all positive instances.

\begin{align*}
   \text{Recall} &= \frac{TP}{TP + FN}
\end{align*}

#### 4. F1 Score

F1 Score is the harmonic mean of precision and recall. It provides a balanced measure between precision and recall, especially when dealing with imbalanced datasets.

\begin{align*}
   \text{F1 Score} &= \frac{TP}{TP + 0.5\cdot(FN + FP)}
\end{align*}


#### 5. Receiver Operator Characteristic (ROC)

The ROC curve shows the performance of a classifier at **all classification thresholds** by plotting the True Positive Rate (TPR) as a function of the False Positive Rate (FPR). The area under this curve (AUC) is a measure of classification performance.

##


\begin{align*}
   \text{TPR} &= \frac{TP}{TP + FN}
\end{align*}


##

\begin{align*}
   \text{FPR} &= \frac{FP}{FP + TN}
\end{align*}

### 9.6 Receiver Operator Characteristic ROC <a id ='9.6' ></a>

Die folgende Funktion wird fÃ¼r das Zeichnen des Random classifiers im ROC-plot benÃ¶tigt

```python
def add_identity(axes, *line_args, **line_kwargs):
    identity, = axes.plot([], [], *line_args, **line_kwargs)
    def callback(axes):
        low_x, high_x = axes.get_xlim()
        low_y, high_y = axes.get_ylim()
        low = max(low_x, low_y)
        high = min(high_x, high_y)
        identity.set_data([low, high], [low, high])
    callback(axes)
    axes.callbacks.connect('xlim_changed', callback)
    axes.callbacks.connect('ylim_changed', callback)
    return axes
```

```python
# Implement a 5-fold stratified crossvalidation loop with relevant data standardization.

# Perform a 5-fold crossvalidation - prepare the splits
n_splits = 5
skf      = StratifiedKFold(n_splits=n_splits)

# Use this counter to save your performance metrics for each crossvalidation fold
# also plot the roc curve for each model and fold into a joint subplot
fold = 0
fig,axs = plt.subplots(2,2,figsize=(12, 9))

# Loop over all splits
for train_index, test_index in skf.split(X_over, y_over):

    # Get the relevant subsets for training and testing
    X_test  = X_over.iloc[test_index]
    y_test  = y_over.iloc[test_index]
    X_train = X_over.iloc[train_index]
    y_train = y_over.iloc[train_index]
    
    #transform y to an 1d array (better for sklearn)
    y_test = column_or_1d(y_test, warn=False)
    y_train = column_or_1d(y_train, warn=False)


    # Standardize the numerical features using training set statistics
    sc = StandardScaler()
    X_train_sc = sc.fit_transform(X_train)
    X_test_sc  = sc.transform(X_test)


    # Creat prediction models and fit them to the training data

    # Logistic regression
    clf_LR = LogisticRegression(random_state=1)
    clf_LR.fit(X_train_sc, y_train)

    # Random forest
    max_depth, max_features, max_leaf_nodes, n_estimators = RF_hyper(X_model, y_model_1d)
    clf_RF = RandomForestClassifier(random_state=1, max_depth = max_depth, max_features = max_features, max_leaf_nodes = max_leaf_nodes, n_estimators = n_estimators)
        
    clf_RF.fit(X_train_sc, y_train)
    
    #Support Vector Maschine
    clf_SVM = svm.SVC(probability=True, kernel = 'linear') # Support Vector Maschine
    clf_SVM.fit(X_train_sc, y_train)

    # K Nearest Neighbors
    clf_KNN = KNeighborsClassifier(n_neighbors= 25) # K-Nearest Neighbors
    clf_KNN.fit(X_train_sc, y_train)

    # Get the roc curve using a sklearn function
    y_test_predict_proba_LR  = clf_LR.predict_proba(X_test_sc)[:, 1]
    fp_rates_LR, tp_rates_LR, _ = roc_curve(y_test, y_test_predict_proba_LR)
    
    y_test_predict_proba_RF  = clf_RF.predict_proba(X_test_sc)[:, 1]
    fp_rates_RF, tp_rates_RF, _ = roc_curve(y_test, y_test_predict_proba_RF)
   
    y_test_predict_proba_SVM  = clf_SVM.predict_proba(X_test_sc)[:, 1]
    fp_rates_SVM, tp_rates_SVM, _ = roc_curve(y_test, y_test_predict_proba_SVM)
   
    y_test_predict_proba_KNN  = clf_KNN.predict_proba(X_test_sc)[:, 1]
    fp_rates_KNN, tp_rates_KNN, _ = roc_curve(y_test, y_test_predict_proba_KNN)
    
    # Plot on the provided axis
    axs.flat[0].plot(fp_rates_LR, tp_rates_LR,label='fold ' + str(fold))
    axs.flat[1].plot(fp_rates_RF, tp_rates_RF,label='fold ' + str(fold))
    axs.flat[2].plot(fp_rates_SVM, tp_rates_SVM,label='fold ' + str(fold))
    axs.flat[3].plot(fp_rates_KNN, tp_rates_KNN,label='fold ' + str(fold))
    
    # increase counter for folds
    fold += 1

# Make plot pretty
model_names = ['LR','RF', 'SVM', 'KNN']
for i,ax in enumerate(axs.flat):
    ax.set_xlabel('FPR')
    ax.set_ylabel('TPR')
    add_identity(ax, color="b", ls="--",label = 'random\nclassifier')
    ax.legend()
    ax.title.set_text('ROC ' +model_names[i])
plt.tight_layout()
```
